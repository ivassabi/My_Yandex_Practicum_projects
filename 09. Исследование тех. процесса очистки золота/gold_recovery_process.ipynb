# Описание данных

**Технологический процесс**

•	Rougher feed — исходное сырье  
•	Rougher additions (или reagent additions) — флотационные реагенты: Xanthate, Sulphate, Depressant  
  o	Xanthate — ксантогенат (промотер, или активатор флотации);  
  o	Sulphate — сульфат (на данном производстве сульфид натрия);  
  o	Depressant — депрессант (силикат натрия).  
•	Rougher process (англ. «грубый процесс») — флотация  
•	Rougher tails — отвальные хвосты  
•	Float banks — флотационная установка  
•	Cleaner process — очистка  
•	Rougher Au — черновой концентрат золота  
•	Final Au — финальный концентрат золота  

**Параметры этапов**

•	air amount — объём воздуха  
•	fluid levels — уровень жидкости  
•	feed size — размер гранул сырья  
•	feed rate — скорость подачи  

**Наименование признаков**

Наименование признаков должно быть такое:  
_[этап].[тип_параметра].[название_параметра]_  
Возможные значения для блока [этап]:  
•	rougher — флотация  
•	primary_cleaner — первичная очистка  
•	secondary_cleaner — вторичная очистка  
•	final — финальные характеристики  

**Возможные значения для блока [тип_параметра]:**  

•	input — параметры сырья  
•	output — параметры продукта  
•	state — параметры, характеризующие текущее состояние этапа  
•	calculation — расчётные характеристики

```python
def correlation(df):

    '''Функция определяет признаки с уровнем корреляции >= 0,9 и формирует список признаков для удаления'''

    col_corr = []
    corr_matrix = df.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if corr_matrix.iloc[i, j] >= 0.9 and (corr_matrix.columns[j] not in col_corr):
                colname = corr_matrix.columns[i] 
                col_corr.append(colname)
                # if colname in df.columns:
                #     del df[colname]
    return col_corr
```

###  Функция для вычисления итоговой sMAPE.

```python
def smape(target, predictions):
    smape = 100/len(target) * sum(2 * abs(predictions - target) / (abs(target) + abs(predictions)))
    return smape
```

```python
def final_smape(smape_rougher, smape_final):
    final_smape = 0.25*smape_rougher+0.75*smape_final
    return final_smape
```

```python
def random(model, params, features, target):
    random = RandomizedSearchCV(model, params, scoring=score).fit(features, target)
    return random.best_params_, random.best_score_
```

```python
def grid(model, params, features, target):
    grid = GridSearchCV(model, params, scoring=score).fit(features, target)
    return grid.best_params_, grid.best_score_
```
